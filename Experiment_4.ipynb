```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Time Series Data:
data = pd.read_csv('data.csv', parse_dates=['Date'], index_col='Date')
```

```python
# Check for the correct column name (replace 'value' if necessary)
# Print the available columns to identify the correct one
print(data.columns)
```

```python
# Replace 'your_actual_column_name' with the actual name of the column from your data
column_name = data.columns[0]
```

```python
# Estimate Trend using Rolling Average:
rolling_mean = data[column_name].rolling(window=12).mean()
plt.figure(figsize=(12, 6))
plt.plot(data[column_name], label='Original Data')
plt.plot(rolling_mean, label='Rolling Mean (Trend)', color='orange')
plt.title('Estimating Trend Using Rolling Average')
plt.xlabel('Date')
plt.ylabel(column_name)  # Update label with the column name
plt.legend()
plt.show()
```

```python

# Eliminate Trend using Differencing:
data['Detrended'] = data[column_name].diff().dropna()
plt.figure(figsize=(12, 6))
plt.plot(data['Detrended'], label='Detrended Data')
plt.title('Time Series After Eliminating Trend')
plt.xlabel('Date')
plt.ylabel(column_name)  # Update label with the column name
plt.legend()
plt.show()
```
