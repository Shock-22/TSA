```python


import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Generate time series data without trend and seasonality
np.random.seed(42)  # for reproducibility
n_samples = 100
data = np.random.normal(loc=0, scale=1, size=n_samples)
time_index = pd.date_range(start='2023-01-01', periods=n_samples, freq='D')
series = pd.Series(data, index=time_index)


# Fit ARIMA model
model = ARIMA(series, order=(5, 1, 0))  # Example order, you might need to adjust
model_fit = model.fit()

# Make predictions
forecast_steps = 10
forecast = model_fit.predict(start=len(series), end=len(series) + forecast_steps - 1)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(series, label='Original Data')
plt.plot(forecast.index, forecast.values, label='Forecast', color='red')
plt.legend()
plt.title('ARIMA Forecast')
plt.show()

forecast
```
